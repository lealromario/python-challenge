```python
import pandas as pd

# Read CSV file
budget_data3 = pd.read_csv(r"C:\Users\lealr\OneDrive\DU Bootcamp\Challenges\Instructions\PyBank\Resources\budget_data.csv")
budget_data3.head()
```

```python
# Counting the number of months
number_months2 = budget_data3["Date"].count
print("Total Months:", number_months2())
```

```python
# The total cash made after adding the profits and losses 
total = budget_data3['Profit/Losses'].sum()
print("Total:$", total)
```

```python
# Using .diff allows me to compare the new values with the previous one
# The round command allows me to round the number up to two decimal places

budget_data3['Average Change'] = budget_data3['Profit/Losses'].diff()
avg_change = budget_data3['Average Change'].mean()
print("Average change:$", round(avg_change, 2))
```

```python
#Finding greatest increase in profits and greatest decrease
budget_data3['Average Change'] = budget_data3['Profit/Losses'].diff()
avg_change_max = budget_data3['Average Change'].max()


budget_data3['Average Change'] = budget_data3['Profit/Losses'].diff()
avg_change_min = budget_data3['Average Change'].min()

print("Greatest Increase in Profits$:",avg_change_max)
print("Greatest Decrease in Loss:$", avg_change_min)
```

```python
# Formatting final statements
print("Financial Analysis")
print("--------------------------------------------------------------------")
print("Total Months:", number_months2())
print("Total:$", total)
print("Average change:$", round(avg_change, 2))
print("Greatest Increase in Profits$:",avg_change_max)
print("Greatest Decrease in Loss:$", avg_change_min)
```
