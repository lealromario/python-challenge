```python
import pandas as pd

# Read the csv file
elections_data = pd.read_csv(r"C:\Users\lealr\OneDrive\DU Bootcamp\Challenges\Instructions\PyPoll\Resources\elections_data.csv")
elections_data
```

```python
# Count the sum of voters
number_of_voters = elections_data["Ballot ID"].count()
print("Total Number of voters: ", number_of_voters)
print("------------------------------------------------------------------------------------------------")
```

```python
# List of candidates who received votes 
# Number of votes each candidate received
# Display winner of general election
vote_counts = elections_data.groupby(['Candidate', 'Ballot ID']).size().reset_index(name='count')
total_votes = vote_counts.groupby('Candidate')['count'].sum()
print(total_votes)
```

```python
vote_percentages = (total_votes / total_votes.sum() *100)
vote_percentages = vote_percentages.map('{:.2f}%'.format)

print(vote_percentages)
```

```python
# using the total_votes dataframe I can determine the winner of the election with the idxmax function
winner = total_votes.idxmax()
print(f"The winner of the election is candidate: {winner}")
```
